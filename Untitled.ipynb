```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Define categorical & numerical columns
categorical_cols = ["Car Name", "Fuel Type", "Seller Type", "Transmission"]
numerical_cols = ["Year", "Present Price", "Kms Driven", "Owner", "Selling Price"]

# LabelEncoders (Assuming already fitted on training data)
label_encoders = {col: LabelEncoder() for col in categorical_cols}

# Fit label encoders on original dataset (df)
for col in categorical_cols:
    df[col] = label_encoders[col].fit_transform(df[col])

# StandardScaler (Assuming already fitted on training numerical data)
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# New data (Raw)
new_data = pd.DataFrame([['Honda Civic', 2016, 11187.87, 78889, 'Diesel', 'Individual', 'Manual', 4, 7253.97]],
                        columns=df.columns)

# Encode categorical columns in new_data
for col in categorical_cols:
    if new_data[col].iloc[0] in label_encoders[col].classes_:
        new_data[col] = label_encoders[col].transform(new_data[col])
    else:
        new_data[col] = -1  # Assign -1 for unseen categories

# Scale numerical columns
new_data[numerical_cols] = scaler.transform(new_data[numerical_cols])

# Convert new_data to DataFrame with original columns
new_data = pd.DataFrame(new_data, columns=df.columns)

# Display transformed new_data
print(new_data)
```
